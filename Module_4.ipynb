```python
import pandas as pd

df_users = pd.read_csv("./data/clean/clean_users.csv", encoding="utf-8-sig")
df_products = pd.read_csv("./data/clean/clean_products.csv", encoding="utf-8-sig")
df_browses = pd.read_csv("./data/clean/clean_browses.csv", encoding="utf-8-sig")
df_orders = pd.read_csv("./data/clean/clean_orders.csv", encoding="utf-8-sig")
print("✅ 清洗数据读取完成！")
```

```python
# 订单表时间处理
df_orders["created_at"] = pd.to_datetime(df_orders["created_at"])
df_orders["order_hour"] = df_orders["created_at"].dt.hour   # 小时特征
df_orders["order_month"] = df_orders["created_at"].dt.month # 月份特征

# 浏览表时间处理
df_browses["created_at"] = pd.to_datetime(df_browses["created_at"])
df_browses["browse_hour"] = df_browses["created_at"].dt.hour

print("✅ 时间特征衍生完成！")
```

```python
# 1. 先删掉 orders 和 products 合并后多余的 id 列
df_merge1 = pd.merge(df_orders, df_products, 
                     left_on="product_id", right_on="id", how="left")
# 删掉来自 products 的 id_y（已用 product_id 关联），
# 以及来自 orders 的 id_x（已作为 order_id 用途）
df_merge1 = df_merge1.drop(columns=["id_x", "id_y", "status_y"])  # status_y 同理可删

# 2. 关联 users 表，删掉多余的 id
df_merge2 = pd.merge(df_merge1, df_users, 
                     left_on="user_id", right_on="id", how="left")
df_merge2 = df_merge2.drop(columns=["id"])  # 删掉来自 users 的 id

# 3. 关联 browses 表，删掉多余的 id
df_full = pd.merge(df_merge2, df_browses,
                   on=["user_id", "product_id"], how="left")
df_full = df_full.drop(columns=["id"])  # 删掉来自 browses 的 id

print(f"融合后宽表数据维度：{df_full.shape}")
print("宽表字段预览：")
print(df_full.columns.tolist())
```
