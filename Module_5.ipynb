```python
import pandas as pd
df = pd.read_csv("./data/feature/full_feature_data.csv", encoding="utf-8-sig")
print("特征宽表读取成功！")
```

```python
# 核心数值字段统计
print("===== 消费与销量核心统计 =====")
print(df[["total_amount", "quantity", "price"]].describe())
```

```python
# 剔除商品信息缺失的行
df_clean = df.dropna(subset=["category_id", "name"]).copy()

# === 商品类目热度统计（category_id）===
print("===== 商品类目订单热度分布 =====")

cate_count = df_clean["category_id"].value_counts().sort_index()
print(f"共 {len(cate_count)} 个类目")
print(f"总订单数: {cate_count.sum():,}")
print()

for cid, cnt in cate_count.items():
    bar = "█" * int(cnt / cate_count.max() * 30)
    print(f"  类目 {int(cid):3d}: {cnt:4d} 单 ({cnt/cate_count.sum()*100:.1f}%)")

print()
print(f"最热门类目: 类目 {int(cate_count.idxmax())}（{cate_count.max()} 单）")
print(f"最冷门类目: 类目 {int(cate_count.idxmin())}（{cate_count.min()} 单）")
```

```python
print("===== 各时段下单量统计 =====")
hour_analyse = df.groupby("order_hour")["user_id"].count().reset_index()
hour_analyse.columns = ["下单小时", "下单数量"]
print(hour_analyse)
```

```python
# 各类目订单量、总销售额、平均客单价
cate_pivot = pd.pivot_table(
    df, index="category_id",
    values=["quantity"],
    aggfunc=["count", "sum", "mean"]
)
print("===== 类目营收交叉分析 =====")
print(cate_pivot)
```
